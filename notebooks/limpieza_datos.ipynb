# Limpieza de valores nulos

Los valores faltantes pueden:

- afectar cálculos
- generar errores
- producir análisis incorrectos

Las estrategias más comunes son:

- eliminar registros
- reemplazar valores
- imputar datos

````python
# Cantidad de valores nulos

df.isnull().sum()
````

````python
# Reemplazo de edades inválidas por NaN

df['edad'] = pd.to_numeric(
    df['edad'],
    errors='coerce'
)

# Imputación usando la mediana

mediana_edad = df['edad'].median()

df['edad'] = df['edad'].fillna(
    mediana_edad
)

df['edad'].head()
````

# Normalización de texto

Los datos categóricos suelen presentar inconsistencias:

```text
Bogotá
bogota
BOGOTA
Bogota
```

Esto afecta agrupaciones y análisis.

````python
# ============================================
# NORMALIZACIÓN AVANZADA DE TEXTO
# ============================================

# Importamos la librería unicodedata
# Esta librería permite trabajar con caracteres Unicode
# y eliminar tildes o caracteres especiales.

import unicodedata


# ============================================
# FUNCIÓN DE LIMPIEZA DE TEXTO
# ============================================

def limpiar_texto(texto):
    
    # Validamos si el valor es nulo
    # Si es nulo, lo retornamos sin modificar
    
    if pd.isnull(texto):
        return texto
    
    
    # Convertimos el valor a string
    # Esto evita errores si existen datos numéricos
    
    texto = str(texto)
    
    
    # ============================================
    # ELIMINACIÓN DE TILDES
    # ============================================
    
    # normalize('NFKD') separa caracteres especiales
    # Ejemplo:
    # Bogotá -> Bogota + tilde separada
    
    texto = unicodedata.normalize(
        'NFKD',
        texto
    )
    
    
    # encode('ascii', 'ignore')
    # elimina caracteres especiales y tildes
    
    texto = texto.encode(
        'ascii',
        'ignore'
    )
    
    
    # decode('utf-8')
    # convierte nuevamente el texto a formato legible
    
    texto = texto.decode(
        'utf-8'
    )
    
    
    # ============================================
    # NORMALIZACIÓN DE FORMATO
    # ============================================
    
    texto = (
        texto
        
        # Convertir todo a minúsculas
        .lower()
        
        # Eliminar espacios al inicio y final
        .strip()
        
        # Capitalizar correctamente
        .title()
    )
    
    
    # Retornar texto limpio
    
    return texto


# ============================================
# APLICAR LIMPIEZA A LA COLUMNA
# ============================================

df['ciudad'] = df['ciudad'].apply(
    limpiar_texto
)


# ============================================
# VALIDAR RESULTADOS
# ============================================

# Mostramos los valores únicos de la columna
# para verificar que ya no existan duplicados
# por tildes o mayúsculas

df['ciudad'].unique()
````

````python
# ============================================
# IMPORTACIÓN DE LIBRERÍAS
# ============================================

import pandas as pd
import numpy as np
import unicodedata
````

# Construcción de funciones reutilizables

En ingeniería de datos se evita repetir código.

Las funciones permiten:

- reutilización
- mantenimiento sencillo
- modularidad
- automatización
- escalabilidad

````python
# ============================================
# FUNCIÓN DE NORMALIZACIÓN DE TEXTO
# ============================================

def limpiar_texto(texto):
    
    # Validar valores nulos
    
    if pd.isnull(texto):
        return texto
    
    
    # Convertir a string
    
    texto = str(texto)
    
    
    # Eliminar tildes
    
    texto = unicodedata.normalize(
        'NFKD',
        texto
    ).encode(
        'ascii',
        'ignore'
    ).decode(
        'utf-8'
    )
    
    
    # Normalización general
    
    texto = (
        texto
        .lower()
        .strip()
        .title()
    )
    
    
    return texto
````

# Automatización de limpieza

Ahora construiremos una función que procese automáticamente el dataset completo.

---

# Objetivos

La función realizará:

- conversión de tipos
- imputación de nulos
- normalización textual
- eliminación de duplicados
- validaciones básicas

````python
# ============================================
# FUNCIÓN PRINCIPAL DEL PIPELINE
# ============================================

def limpiar_dataset(df):
    
    # Copia de seguridad
    
    df = df.copy()
    
    
    # ============================================
    # LIMPIEZA DE EDAD
    # ============================================
    
    df['edad'] = pd.to_numeric(
        df['edad'],
        errors='coerce'
    )
    
    
    # Imputación usando mediana
    
    mediana_edad = df['edad'].median()
    
    df['edad'] = df['edad'].fillna(
        mediana_edad
    )
    
    
    # ============================================
    # LIMPIEZA DE PROMEDIO
    # ============================================
    
    df['promedio'] = pd.to_numeric(
        df['promedio'],
        errors='coerce'
    )
    
    
    promedio_mediana = df['promedio'].median()
    
    df['promedio'] = df['promedio'].fillna(
        promedio_mediana
    )
    
    
    # ============================================
    # NORMALIZACIÓN DE CIUDAD
    # ============================================
    
    df['ciudad'] = df['ciudad'].apply(
        limpiar_texto
    )
    
    
    # ============================================
    # ELIMINACIÓN DE DUPLICADOS
    # ============================================
    
    df = df.drop_duplicates()
    
    
    # ============================================
    # FILTRADO DE OUTLIERS
    # ============================================
    
    df = df[
        (df['edad'] >= 15) &
        (df['edad'] <= 90)
    ]
    
    
    return df
````

# Ejecución del pipeline

Ahora ejecutaremos el proceso automatizado sobre el dataset.

````python
# ============================================
# EJECUCIÓN DEL PIPELINE
# ============================================

df_limpio = limpiar_dataset(df)

df_limpio.head()
````
