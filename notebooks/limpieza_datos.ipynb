# Problema de tipos de datos

En datasets reales es común encontrar columnas numéricas con valores incorrectos:

```text
25
30
treinta
ciento veinte
```

Esto genera errores al realizar comparaciones o cálculos.

Pandas permite convertir valores utilizando:

```python
pd.to_numeric()
```

Los valores inválidos pueden convertirse automáticamente en `NaN`.

````python
df[df['edad'] > 30].head()
````

````python
# Conversión temporal de la columna edad a numérico

edad_numerica = pd.to_numeric(
    df['edad'],
    errors='coerce'
)

# Filtrar mayores de 30 años

df[edad_numerica > 30].head()
````

# Conteo de valores únicos

````python
df['ciudad'].value_counts()
````
